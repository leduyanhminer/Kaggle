```python
import torch
import matplotlib.pyplot as plt 
from torch import nn
import pandas as pd

from datetime import  datetime
```

```python
data_frame = pd.read_csv('dataset/train.csv', index_col="id")
data_frame.head()
```

```python
# Chuẩn hoá một số cột dữ liệu về đoạn [0,1]
from sklearn.preprocessing import MinMaxScaler
def normalize_dataframe(df, columns):
    eps = 0.001
    scaler = MinMaxScaler()
    df[columns] = scaler.fit_transform(df[columns])
    df[columns] += eps
    return df
```

```python
columns_to_normalize = ['age', 'height(cm)', 'weight(kg)', 'waist(cm)', 'systolic', 'relaxation', 'fasting blood sugar', 'Cholesterol', 
                        'triglyceride', 'HDL', 'LDL', 'hemoglobin', 'AST', 'ALT', 'Gtp']
```

```python
data_frame = normalize_dataframe(data_frame, columns_to_normalize)
data_frame
```

```python
data_frame.info()
```

```python

y = data_frame["smoking"]
x = data_frame.drop(columns="smoking")
x = torch.tensor(x.values, dtype=torch.float)
y = torch.tensor(y.values, dtype=torch.float)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
model = nn.Sequential(
    nn.Linear(in_features = 22, out_features = 128),
    nn.ReLU(),
    nn.Linear(in_features = 128, out_features = 64),
    nn.ReLU(),
    nn.Dropout(p=0.5),
    nn.Linear(in_features = 64, out_features = 1)
)

cur_time = datetime.now().today().strftime("%d_%m_%Y")
MODEL_SAVE_PATH = f"model\smoke_status_model_{cur_time}.pth"
PRETRAINED_PATH = r"D:\Kaggle\SmokeStatus\model\smoke_status_model_01.pth"
model.load_state_dict(torch.load(PRETRAINED_PATH), strict=False)
```

```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(params=model.parameters(), lr = 0.01)
```

```python
def accuracy_fn(y_pred, y_true):
    correct = torch.eq(y_true, y_pred).sum().item()
    acc = (correct / len(y_pred)) * 100
    return acc
```

```python
epochs = 500
epoch_count = []
train_loss_value = []
test_loss_value = []
for epoch in range(epochs):
    model.train()
    y_logits = model(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits))
    loss = loss_fn(y_logits, y_train)
    acc = accuracy_fn(y_pred, y_train)
    train_loss_value.append(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()                     
    model.eval()
    with torch.inference_mode():
        test_logits = model(X_test).squeeze()
        test_pred = torch.round(torch.sigmoid(test_logits))
        test_loss = loss_fn(test_logits, y_test)
        test_acc = accuracy_fn(test_pred, y_test)
        test_loss_value.append(test_loss)
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Train_acc: {acc:.5f}, Test_acc: {test_acc:.5f}")
```

```python
torch.save(obj = model.state_dict(), f=MODEL_SAVE_PATH)
```
